**Stage 1 :** Collect data and preprocessing

First delete unwanted column , then drop a no value rows , then copy the cleaned data into new variable and create a new variable to convert binary value of the variable . threshold 24.99 above is high humidity

```python
#check before clean the rows
before_rows = data.shape[0]
print(before_rows)
```

```python
data = data.dropna()

#check after rows
after_rows = data.shape[0]
print(after_rows)
```

```python
cleaned_data = data.copy()
cleaned_data['binary_y'] = (cleaned_data['relative_humidity_3pm'] > 24.99)*1
cleaned_data['binary_y'].head()
```

**Stage2 :**  Training

First we initialize the Y and then create a variable morning_weather put all columns except Y ... then Split the X and Y for test and train data using train_test_split function . then create a variable for decisionclassifier training object , then give the test data

```python
Y = cleaned_data['binary_y'].copy()
```

```python
morning_feature = ['air_pressure_9am', 'air_temp_9am', 'avg_wind_direction_9am',
       'avg_wind_speed_9am', 'max_wind_direction_9am', 'max_wind_speed_9am',
       'rain_accumulation_9am', 'rain_duration_9am', 'relative_humidity_9am']

X = cleaned_data[morning_feature].copy()
X.columns
```

```python
# now we can split the data for train and test
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.30 , random_state=280 )
```

```python
# now we create a classifier object and train the test data
humidity_3pm = DecisionTreeClassifier(max_leaf_nodes=17, random_state=0 , criterion="gini")

humidity_3pm.fit(X_train,Y_train)
```

**Stage 3 :** Prediction

Now we on the prediction stage . we have a trained model humidity 3pm . create a variable for model then check . then compare with actual ytest

```python
Y_predicted = humidity_3pm.predict(X_test)

Y_predicted[:10]
```

```python
Y_test[:10]
```

Finally we can check the accuracy of our model using accuracy score function

```python
accuracy_score(Y_test, Y_predicted)*100
```
